## Initializing And Loading Dataset From CSV file "data.csv"

```python
import csv
import random



# classroom_rooms and large_hall_rooms 
classroom_rooms = ["Room1", "Room2", "Room3", "Room4", "Room5", "Room6", "Room7", "Room8", "Room9", "Room10"]
large_hall_rooms = ["Room11", "Room12", "Room13", "Room14", "Room15", "Room16", "Room17", "Room18", "Room19", "Room20"]

class TableSlot:
    def __init__(self, professor, batch, course, section, strength, room, timeslot, day):
        self.professor = professor
        self.batch = batch
        self.course = course
        self.section = section
        self.strength = strength
        
        if self.strength > 60:
            self.room = self.assign_large_hall()
        else:
            self.room = self.assign_classroom()

        self.timeslot = timeslot
        self.day = day
        
    def assign_classroom(self):
        return  random.choice(classroom_rooms)
    
    def assign_large_hall(self):
        return  random.choice(large_hall_rooms)

professor_list = []
batch_list = []
course_list = []
section_list = []
strength_list = []
room_list = []
timeslot_list = []
day_list = []

with open('data.csv', 'r') as file:
    csv_reader = csv.DictReader(file)
    for row in csv_reader:
        professor_list.append(row['Professor'])
        batch_list.append(row['Batch'])
        course_list.append(row['Course'])
        section_list.append(row['Section'])
        room_list.append(row['Room'])
        timeslot_list.append(row['Time Slot'])
        day_list.append(row['Days'])
```

### Remove empty values from lists

```python
# Remove empty values 
professor_list = [professor for professor in professor_list if professor.strip()]
batch_list = [batch for batch in batch_list if batch.strip()]
course_list = [course for course in course_list if course.strip()]
section_list = [section for section in section_list if section.strip()]
strength_list = [random.randint(40, 80) for _ in range(len(section_list)*len(batch_list))]
room_list = [room for room in room_list if room.strip()]
timeslot_list = [timeslot for timeslot in timeslot_list if timeslot.strip()]
day_list = [day for day in day_list if day.strip()]



# You can now access the data stored in these lists
print("Professors:", professor_list)
print("Batches:", batch_list)
print("Courses:", course_list)
print("Sections:", section_list)
print("Strengths:", strength_list)
print("Rooms:", room_list)
print("Timeslots:", timeslot_list)
print("Days:", day_list)
```

## Genetic Algorithm with following steps: reproduction, crossover, and mutation.

```python




total_batches = len(batch_list)
total_sections = len(section_list)
total_days = len(day_list)
total_timeslot= len(timeslot_list)
total_possible_slots=total_batches*total_sections*total_timeslot*2




timetables = []  # Initialize 



def fitness_function(timetable):
    fitness = 0
    
    
     # Constraint : Classes can only be scheduled in free classrooms.
     # Constraint : Classroom capacity constraint
     # Constraint : Professor should not be assigned two different lectures at the same time slot.
     # Constraint : Same course and section cannot be assigned to two different rooms at the same time slot.
     # Constraint : No professor can teach more than 3 courses but different section.
     # Constraint : No section can have more than 5 courses in a semester.
     # Constraint : Each course would have two lectures per week not on the same or adjacent days.
     # Constraint : Lab lectures should be conducted in two consecutive slots
    
    # Constraint 1
    occupied_rooms = {}
    for slot in timetable:
        key = (slot.room, slot.day, slot.timeslot)
        if key in occupied_rooms:
            fitness -= 1  
        else:
            occupied_rooms[key] = True
    
    # Constraint 2
    for slot in timetable:
        if slot.room in classroom_rooms:  
            if slot.strength > 60:
                fitness -= 1  
        elif slot.room in large_hall_rooms:  
            if slot.strength > 120:
                fitness -= 1  
    
    # Constraint 3
    professor_timeslots = {}
    for slot in timetable:
        key = (slot.professor, slot.day, slot.timeslot, slot.course)
        if key in professor_timeslots:
            fitness -= 1  
        else:
            professor_timeslots[key] = True
    
    # Constraint 4
    course_timeslots = {}
    for slot in timetable:
        key = (slot.course, slot.section, slot.day, slot.timeslot,slot.batch)
        if key in course_timeslots:
            fitness -= 1 
        else:
            course_timeslots[key] = True
    
    # Constraint 5
    professor_courses = {}
    for slot in timetable:
        key = (slot.professor, slot.course)
        if key in professor_courses:
            professor_courses[key].add(slot.section)
        else:
            professor_courses[key] = {slot.section}
    
    for sections in professor_courses.values():
        if len(sections) > 3:
            fitness -= 1  
    
    # Constraint 6
    section_courses = {}
    for slot in timetable:
        key = (slot.section, slot.course)
        if key in section_courses:
            section_courses[key] += 1
        else:
            section_courses[key] = 1
    
    for count in section_courses.values():
        if count > 5:
            fitness -= 1  
    
    # Constraint 7
    course_lecture_days = {}
    for slot in timetable:
        key = (slot.batch,slot.course, slot.section)
        if key in course_lecture_days:
            course_lecture_days[key].append(int(slot.day[3:]))  
        else:
            course_lecture_days[key] = [int(slot.day[3:])]
    
    for days in course_lecture_days.values():
        days.sort()
        for i in range(len(days) - 1):
            if days[i+1] == days[i] + 1:  
                fitness -= 1  
    
    # Constraint 8
    for i in range(len(timetable) - 1):
        if timetable[i].course == course_list[4] and timetable[i+1].course == course_list[4] and timetable[i].timeslot == timeslot_list[4] and timetable[i+1].timeslot == timeslot_list[5]:
            continue
        elif timetable[i].course == course_list[4] and timetable[i+1].course == course_list[4] and timetable[i].timeslot == timeslot_list[5] and timetable[i+1].timeslot == timeslot_list[4]:
            continue
        elif timetable[i].course == course_list[4] and timetable[i+1].course == course_list[4] and timetable[i].timeslot ==  timetable[i+1].timeslot :
            continue
        elif timetable[i].course == course_list[4]:
            fitness -= 1  
    print("Fitness Score : ",fitness)
    return fitness




def initialize_population():
    population = []
    for _ in range(100):
        timetable = []
        for _ in range(total_possible_slots):
            slot = TableSlot(
                random.choice(professor_list),
                random.choice(batch_list),
                random.choice(course_list),
                random.choice(section_list),
                random.choice(strength_list),
                random.choice(room_list),
                random.choice(timeslot_list),
                random.choice(day_list)
            )
            timetable.append(slot)
        population.append(timetable)
    return population

def evaluate_fitness(population):
    fitness_scores = []
    for timetable in population:
        
        fitness_score = fitness_function(timetable)
        fitness_scores.append(fitness_score)
    return fitness_scores

def reproduce(parent1, parent2):
    offspring = []
    for i in range(len(parent1)):
        #crossover
        if random.random() < 0.5:
            offspring.append(parent1[i])
        else:
            offspring.append(parent2[i])

            #mutation
    if random.random() < 0.01:
        i = random.randint(0, len(offspring) - 1)
        offspring[i] = TableSlot(
            random.choice(professor_list),
            random.choice(batch_list),
            random.choice(course_list),
            random.choice(section_list),
            random.choice(strength_list),
            random.choice(room_list),
            random.choice(timeslot_list),
            random.choice(day_list)
        )

    return offspring

def selection(population, fitness_scores):
    fittest_timetables = [timetable for _, timetable in sorted(zip(fitness_scores, population), key=lambda pair: pair[0], reverse=True)]
    return fittest_timetables[:1]

def genetic_algorithm():
    population = initialize_population()
    for _ in range(1000):
        fitness_scores = evaluate_fitness(population)
        population = selection(population, fitness_scores)
        new_population = []
        for _ in range(50):
            parent1 = random.choice(population)
            parent2 = random.choice(population)
            offspring = reproduce(parent1, parent2)
            new_population.append(offspring)
        population = new_population

    fittest_timetable = max(population, key=fitness_function)
    return fittest_timetable






fittest_timetable = genetic_algorithm()

fittest_fitness_score = fitness_function(fittest_timetable)




binary_chromosome = display_binary_chromosome(fittest_timetable)
print(binary_chromosome)

fittest_timeetable = decode_binary_chromosome(binary_chromosome)


def group_by_day(slots):
    days = {}
    for slot in slots:
        if slot.day not in days:
            days[slot.day] = []
        days[slot.day].append(slot)
    return days

timetable_by_day = group_by_day(fittest_timetable)

def display_timetable_for_day(day, slots):
    print(f"Day {day}:")
    for i, slot in enumerate(slots):
        print(f"Slot : {i}, Professor: {slot.professor}, Batch: {slot.batch}, Course: {slot.course}, Section: {slot.section}, Room: {slot.room}, Time Slot: {slot.timeslot}, Day: {slot.day}")
#        print(f"Slot : {i}, Professor: {slot.professor}, Batch: {slot.batch}, Course: {slot.course}, Section: {slot.section}, Strength: {slot.strength}, Room: {slot.room}, Time Slot: {slot.timeslot}, Day: {slot.day}")

        
for day, slots in sorted(timetable_by_day.items()):
    display_timetable_for_day(day, slots)

print("Fitness Score:", fittest_fitness_score)

```
